По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые — получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма. 


Таблица с данными состоит из следующих колонок:

* `'airports_nearest'` — расстояние до ближайшего аэропорта в метрах (м)
* `'balcony'` — число балконов
* `'ceiling_height'` — высота потолков (м)
* `'cityCenters_nearest'` — расстояние до центра города (м)
* `'days_exposition'` — сколько дней было размещено объявление (от публикации до снятия)
* `'first_day_exposition'` — дата публикации
* `'floor'` — этаж
* `'floors_total'` — всего этажей в доме
* `'is_apartment'` — апартаменты (булев тип)
* `'kitchen_area'` — площадь кухни в квадратных метрах (м²)
* `'last_price'` — цена на момент снятия с публикации
* `'living_area'` — жилая площадь в квадратных метрах (м²)
* `'locality_name'` — название населённого пункта
* `'open_plan'` — свободная планировка (булев тип)
* `'parks_around3000'` — число парков в радиусе 3 км
* `'parks_nearest'` — расстояние до ближайшего парка (м)
* `'ponds_around3000'` — число водоёмов в радиусе 3 км
* `'ponds_nearest'` — расстояние до ближайшего водоёма (м)
* `'rooms'` — число комнат
* `'studio'` — квартира-студия (булев тип)
* `'total_area'` — площадь квартиры в квадратных метрах (м²)
* `'total_images'` — число фотографий квартиры в объявлении

#### Заполнение пропущенных значений в колонке `'floors_total'`

Выведим первые десять строк с пропусками в колонке `'floors_total'`.

```python
df.loc[df['floors_total'].isna()].head(10)
```

Предположение о том, что значения в колонке `'floors_total'` не заполнены, потому что в доме только один этаж, не подтвердилось. Соответствующие значения в колонке `'floor'` равны не только `1`. 


Посчитаем количество пропусков в колонке.

```python
len(df[df['floors_total'].isna()])
```

Количество пропусков в колонке `'floors_total'` незначительно по отношению ко всем данным. Определим медианное значение для данной колонки.

```python
df['floors_total'].median()
```

Медианное значение - `9` этажей. Дом такой высоты можно втретить в большинстве городов и их районов. Но пропущенным значениям могут соответствовать значение столбца `'floor'` больше `9` и тут медиана не подойдет. Напишем цикл, который поможет этого избежать.

```python
for value in df.loc[df['floors_total'].isna(), 'floors_total']:
  for indx in df.loc[df['floors_total'].isna(), 'floor']:
    if indx <= df['floors_total'].median():
      (
          df.loc[(df['floors_total'].isna()) & (df['floor'] == indx),
                 'floors_total']
       ) = df['floors_total'].median()
       
    elif indx > df['floors_total'].median():
      df.loc[(df['floors_total'].isna()) & (df['floor'] == indx), 'floors_total'] = indx
    
df['floors_total'].isna().sum() # проверяем, все ли пропуски заполнены.
```

##### Исправление неявных дубликатов в колонке `'locality_name'`

Прежде, чем работать над заполнением пропусков в других колонках, необходимо устранить неявные дубликаты в `'locality_name'`. Для начала выведим уникальные значения в этой колонке.

```python
df['locality_name'].unique()
```

Дубликатов много. Например, есть `поселок Мурино`, `посёлок Мурино` и `Мурино`. Составим функцию, которая будет удалять лишние слова и, по-максимуму, оставлять только названия населенных пунктов.

Сначала выведим десятку городов с наибольшим количеством объявлений, затем применим функцию и результаты запишем в новой колонке. После - снова выведим эту десятку, чтобы посмотреть, как она изменится.

```python
df['locality_name'].value_counts().head(10)
```

```python
def loc_name(value):
    name = ['посёлок городского типа ', 'поселок городского типа ', 'деревня ',
            'село ', 'городской поселок ', 'городской посёлок ', 'посёлок ',
            'поселок ']
    for n in name:
        if n in value:
            value = value.replace(n, '')
        else:
            value = value
    return value


df['locality_name_new'] = df['locality_name'].apply(loc_name)
```

```python
df.head() # посмотрим, как выглядит таблица с новой колонкой
```

```python
df['locality_name_new'].value_counts().head(10)
```

```python
df['locality_name_new'].unique() # посмотрим уникальные значения новой колонки
```

Десятка городов с наибольшим количеством объявлений немного изменилась. `Кудрово` поднялось на третью строчку и подвинуло остальные города.

В новой колонке остались указания `садовое товарищество` и т.п., но среди них дубликатов нет, поэтому оставляем так.

## Расчёты и добавление результатов в таблицу

### Расчет цены квадратного метра

Чтобы рассчитать стоимость квадратного метра, нужно стоимость разделить на общую площадь. Результат вычисления запишем в новой колонке `'one_meter_price'`.

```python
df['one_meter_price'] = df['last_price'] / df['total_area']
df['one_meter_price'] = df['one_meter_price'].astype('int')

df.head() #посмотрим, как выглядит таблица с новой колонкой
```

### Расчет соотношения жилой и общей площади

Расчитаем отношение жилой площади к общей площади. Результат вычисления запишем в новой колонке `'living_to_total_area'`.

```python
df['living_to_total_area'] = (df['living_area'] / df['total_area']).round(2)

df.head() #посмотрим, как выглядит таблица с новой колонкой
```

### Расчет соотношения площади кухни и общей площади

Расчитаем отношение площади кухни к общей площади. Результат вычисления запишем в новой колонке `'kitchen_to_total_area'`.

```python
df['kitchen_to_total_area'] = (df['kitchen_area'] / df['total_area']).round(2)

df.head() #посмотрим, как выглядит таблица с новой колонкой
```

### Добавление колонки с категориями по этажам

Сформируем три категории в зависимости от того, на каком этаже находится объект недвижимости:
* `первый` - для первого этажа;
* `последний` - для последнего этажа;
* `другой` - для остальных этажей.

```python
def floor_category(floor):
    if floor == 1:
        return 'первый'
    else:
        return 'другой'

df['floor_category'] = df['floor'].apply(floor_category)

df.loc[df['floor'] == df['floors_total'], 'floor_category'] = 'последний'


df.head(20) # посмотрим, как выглядит таблица с новой колонкой
```
